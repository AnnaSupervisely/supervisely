```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from fast_inference import ICNetFastApplier
from os.path import join
```

```python
images_path = '/sly_task_data/data/' # it could be one image or directory with several images
```

```python
# inference params
settings = {
  "device_id": 0,
  "gpu_devices": [
    0
  ],
  "model_classes": {
    "save_classes": "__all__",
    "add_suffix": "_icnet"
  },
  "existing_objects": {
    "save_classes": [],
    "add_suffix": ""
  },
  "mode": {
    "source": "full_image"
  }
}
```

```python
applier = ICNetFastApplier(settings)
```

```python
def get_imgs_list(path):
    if os.path.isdir(path):
        imgs_list = [join(path, fp) for fp in os.listdir(path)]
    else:
        imgs_list = [path]
    return imgs_list
```

```python
# Get images list
imgs_list = get_imgs_list(images_path)
```

```python
color_mapping = dict()
for cls in applier.train_classes._classes_lst:
    color_mapping[cls['title']] = cls['color']
```

```python
for img_fp in imgs_list:
    plt.figure(figsize=(50, 50))
    image = cv2.imread(img_fp)
    if image is None:
        print('Something wrong with image: {}'.format(img_fp))
        break
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.subplot(1,2,1)
    plt.imshow(image)
    mask = applier.inference(image)
    for obj in mask["objects"]:
        if "person" in obj.class_title:
            obj.draw(image, color_mapping[obj.class_title])
    plt.subplot(1,2,2)
    plt.imshow(image)
    plt.show()
```
